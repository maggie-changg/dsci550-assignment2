Similarly did imagecaptions.csv

```python
import pandas as pd
objects_df = pd.read_csv('objectdetection.csv')
```

```python
duplicate_rows = objects_df[objects_df.duplicated()]
duplicate_rows
```

```python
captions_df = pd.read_csv('imagecaptions.csv')
```

```python
merged_df = pd.concat([captions_df, objects_df], axis=1)
```

```python
merged_df
```

```python
merged_df = merged_df.loc[:, ~merged_df.columns.duplicated()]
```

```python
merged_df.to_csv('objects_captions.csv', index=False)
```

```python
haunted_places_df = pd.read_csv('Haunted Places Final.tsv', sep='\t')
haunted_places_df
```

```python
def make_safe_filename(text, max_length=50):
    safe_text = ''.join(c if c.isalnum() or c in (' ', '_') else '_' for c in text)
    return safe_text[:max_length].strip().replace(' ', '_') + ".png"

haunted_places_df['Safe_Filename'] = haunted_places_df['description'].apply(make_safe_filename)
```

```python
haunted_places_df['Safe_Filename']
```

```python
matching_records = haunted_places_df[haunted_places_df['Safe_Filename'].isin(merged_df['Filename'])]
matching_records
```

```python
haunted_places_df
```

```python
objcap_df = pd.read_csv('objects_captions.csv')
```

```python
merged_haunted_objcap_df = haunted_places_df.merge(objcap_df, left_on='Safe_Filename', right_on='Filename', how='inner')
merged_haunted_objcap_df
```

```python
merged_haunted_objcap_df = merged_haunted_objcap_df.drop(columns=['Safe_Filename'])
merged_haunted_objcap_df
```

```python
merged_haunted_objcap_df.to_csv('merged_haunted_objcap.csv', index=False)
```
