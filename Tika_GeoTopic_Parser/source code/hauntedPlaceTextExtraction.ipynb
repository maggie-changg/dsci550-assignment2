```python
import pandas as pd
import os
```

```python
df = pd.read_csv('../../data/haunted_places_entities.tsv', sep='\t', engine='python')
```

```python
df
```

```python
# create the folder to store the .txt files
if not os.path.exists("Haunted_Places_Text_Files"):
    os.mkdir("../../Tika_GeoTopic_Parser/Haunted_Places_Text_Files")
```

```python
# Iterate over each row and write the description, city, and country to a new .txt file
for index, row in df.iterrows():
    description = str(row["description"])
    location = str(row["location"])
    
    filename = f"Haunted_Places_{index}.txt"
    filepath = f"../../Tika_GeoTopic_Parser/Haunted_Places_Text_Files/{filename}"

    # Write content to the file
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(f"Location: {location}\n\n")
        f.write(f"Description:\n{description}")

        

# Iterate over each row and write the enriched description (with city/country) to a new .txt file
# for index, row in df.iterrows():
#     description = str(row["description"]).strip()
#     city = str(row["city"]).strip()
#     country = str(row["country"]).strip()

#     # Embed the city + country into the start of the description
#     enriched_description = f"{city}, {country}. {description}"

#     filename = f"Haunted_Places_{index}.txt"
#     filepath = f"TEST_Haunted_Places_Text_Files/{filename}"

#     with open(filepath, "w", encoding="utf-8") as f:
#         f.write(enriched_description)
```
